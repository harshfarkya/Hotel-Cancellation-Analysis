### Data Analysis And Visualization

```python
cancelled_prec = df['is_canceled'].value_counts(normalize = True)*100
print(cancelled_prec)

plt.figure(figsize = (5,4))
plt.title("Reservation Status Count")
plt.bar(['Not Canceled','Canceled'],df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (8,4))

ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df , palette = 'Blues')
ax1
plt.title("Reservation Status in Different hotels", size = 20)
plt.xlabel('hotel')
plt.ylabel('Number of Reservation')
plt.show()
```

```python
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)*100
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)*100
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort hotel', size = 20)
plt.plot(resort_hotel.index,resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (20,9))
ax1 = sns.countplot(x = 'month' , hue = 'is_canceled', data = df)
plt.title("Revenue Status Per Month",size = 29)
ax1.legend(['Not Canceled', 'Canceled'])
plt.show()
```

```python
plt.figure(figsize=(20, 8))
plt.title('Avg DR by Month', fontsize=20)
cancelled_reservations = df[df['is_canceled'] == 1]
avg_adr_by_month = cancelled_reservations.groupby('month')[['adr']].sum().reset_index()
sns.barplot(x='month', y='adr', data=avg_adr_by_month)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Average Daily Rate (ADR)', fontsize=14)
plt.show()
```

```python
canceled_data = df[df['is_canceled']==1]
top_10_country = canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8, 8))
plt.title("Top 10 Country By Revservation Canceled")
plt.pie(top_10_country,autopct = '%.2f',labels = top_10_country.index)
plt.show()
```
